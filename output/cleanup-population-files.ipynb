```python
import pandas as pd
```

During the generation process, many columns were added that are redundant. In order to make the synthetic population file clearer, it is best to clean up those columns

# Individuals

```python
df_synth_pop = pd.read_pickle('synthetic_population/with_households/individuals/synth_pop_DHWZ_v10.pkl')
df_synth_pop
```

```python
df_synth_pop.drop(columns=[
    'age_group',
    'small_age_group',
    'education_attainment_age_group',
    'education_age_group',
    'license_age',
], inplace=True)
df_synth_pop
```

```python
df_synth_pop.to_csv('synthetic_population_the_hague_south_west.csv')
```

# Households

```python
df_synth_hh = pd.read_pickle('synthetic_population/with_households/households/synth_households_DHWZ_v10.pkl')
df_synth_hh
```

```python
df_synth_hh.drop(columns=[
    'hh_type',
    'hh_size',
    'small_hh_type',
    'main_bread_winner_age',
    'main_bread_winner_migration_background',
    'income_age_group',
    'income_household_type',
    'car_license',
    'motorcycle_license',
    'vehicle_ownership_income_group'
], inplace=True)
df_synth_hh
```

```python
df_synth_hh.to_csv('synthetic_households_the_hague_south_west.csv')
```
